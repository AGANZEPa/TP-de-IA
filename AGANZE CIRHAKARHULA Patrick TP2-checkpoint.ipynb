# COMPRENDRE LES CONCEPTS DE RIDGE ET DE LASSO

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
```

### On entre les données (DATA SET) Exemple de diabete

```python
diabetes = load_diabetes()
X = diabetes.data
Y = diabetes.target
```

### separation des données en ensemble d'entrainement et de test

```python

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)

# Initialisation des modeles Ridge et LASSO

ridge_model = Ridge(alpha = 1.0)
lasso_model = Lasso(alpha = 1.0)
```

### Entrainement de modeles et predictions

```python

ridge_model.fit (X_train, Y_train)
lasso_model.fit (X_train, Y_train)

ridge_predictions = ridge_model.predict (X_test)
lasso_predictions = lasso_model.predict (X_test)
```

### evaluation des performances et print resultats

```python

ridge_mse = mean_squared_error (Y_test, ridge_predictions)
lasso_mse = mean_squared_error (Y_test, lasso_predictions)

ridge_r2 = r2_score (Y_test, ridge_predictions)
lasso_r2 = r2_score (Y_test, lasso_predictions)

print ("Ridge regression est ")
print (f"Mean Squared Error:{ridge_mse:.2f}")
print (f"R² Score: {ridge_r2:.2f}")

print ("\nLASSO Regression :")
print (f"Mean Squared Error:{lasso_mse:.2f}")
print (f"R² Score: {lasso_r2:.2f}")
```

### presentation des resultats

```python
plt.figure (figsize = (12, 6))

plt.scatter (Y_test, ridge_predictions, color = "green", label = "Ridge Predictions", alpha = 0.5)
plt.scatter (Y_test, lasso_predictions, color = "red", label = "LASSO Predictions", alpha = 0.5)

# La droite d'identité
plt.plot([Y.min(), Y.max()], [Y.min(), Y.max()], color = "blue") 
#......................................................................................................

plt.xlabel(" True values")
plt.ylabel(" Predictions")
plt.title("RIDGE et LASSO predictions sur les diabetes Dataset", color = "red")

plt.legend()
plt.show()
```
